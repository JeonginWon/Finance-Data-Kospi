```python
#변수 각자 확인
import FinanceDataReader as fdr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%config inlinebackend.figure_formats = {'png', 'retina'}
```

```python
#kospi주가
kospi = fdr.DataReader("KS11","2015")
kospi.tail(2)
```

```python
plt.figure(figsize =(20,5))
plt.plot(kospi.index, kospi.Close) #X=날짜, Y=종가
plt.show()
```

```python
#kosdaq주가
kosdaq = fdr.DataReader("KQ11",'2015')
plt.figure(figsize=(20,5))
plt.plot(kosdaq.index, kosdaq.Close)
plt.show()
```

```python
#원달러 환율
usd = fdr.DataReader('USD/KRW','2015')
plt.figure(figsize =(20,5))
plt.plot(usd.index, usd.Close)
plt.show()
```

코스피와 코스닥은 양의 상관관계,
이와 원달러는 음의 상관관계로 보인다
그래프를 합쳐보면 아래와 같음.

```python
#코스피,코스닥,원달러환율 데이터 시각화
plt.figure(figsize=(20,5))
plt.plot(kospi.index, kospi.Close,label='kospi')
plt.plot(kosdaq.index, kosdaq.Close, label='kosdaq')
plt.plot(usd.index, usd.Close, label='usd')
plt.legend()
plt.show()
```

```python
#min max scaler >> 최소값을 0 최대값을 1로 데이터 변환(=정규화)
#(x_1 -min(x))/ (max(x)-min(x))
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
minmax_scale(usd.Close)
plt.figure(figsize=(20,5))
plt.plot(kospi.index, minmax_scale(kospi.Close), label='kospi')
plt.plot(kosdaq.index,minmax_scale(kosdaq.Close), label='kosdaq')
plt.plot(usd.index, minmax_scale(usd.Close), label='usd')
plt.legend()
plt.show()
```

```python
##상관계수
len(kospi), len(kosdaq), len(usd)
```

```python
f_kospi = kospi.reset_index()[['Date','Close']]
f_usd = usd.reset_index()[['Date','Close']]
```

```python
#merge => how='inner', on='date'
merged_df = pd.merge(f_kospi, f_usd, on='Date')
#왼쪽을 코스피, 오른쪽으로 us, 비교하기로 한 주제인 기간을 기준 
merged_df = merged_df.rename(columns ={"Close_x": 'kospi','Close_y':'usd'})#칼럼명 바꾸기
merged_df['kosdaq']= kosdaq.Close.values
merged_df.tail(2)
```

```python
corr_df= np.round(merged_df[['kospi','kosdaq','usd']].corr(),2)
corr_df
```

```python
#상관계수 시각화
#heatmap
import seaborn as sns
sns.heatmap(corr_df, annot=True)
```
